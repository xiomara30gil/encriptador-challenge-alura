```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq

# Parámetros
fs = 500000  # Frecuencia de muestreo (mayor que 2x la máx. frecuencia)
t = np.linspace(0, 1e-3, fs, endpoint=False)  # 1 ms de señal

# Solicita datos al usuario
A_amp = float(input("Ingrese la amplitud de la señal A (diente de sierra): "))
A_freq = float(input("Ingrese la frecuencia de la señal A (Hz, máx. 100kHz): "))
B_amp = float(input("Ingrese la amplitud de la señal B (sen + cos): "))
B_freq = float(input("Ingrese la frecuencia de la señal B (Hz, máx. 100kHz): "))

# Asegura que las frecuencias no excedan 100 kHz
A_freq = min(A_freq, 100000)
B_freq = min(B_freq, 100000)

# Define las señales
signal_A = A_amp * 2 * (t * A_freq - np.floor(0.5 + t * A_freq))  # Diente de sierra
signal_B = B_amp * (np.sin(2 * np.pi * B_freq * t) + np.cos(10 * np.pi * B_freq * t))  # sen + cos

# Grafica señales
def plot_signal(signal, title):
    plt.figure(figsize=(10, 4))
    plt.plot(t[:1000], signal[:1000])  # Mostrar solo primeros 1000 puntos
    plt.title(title)
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Amplitud")
    plt.grid()
    plt.show()

plot_signal(signal_A, "Señal A - Diente de Sierra")
plot_signal(signal_B, "Señal B - sen + cos")

# Pregunta si desea ver la FFT
fft_option = input("¿Desea ver la FFT de las señales? (s/n): ").strip().lower()
if fft_option == 's':
    def plot_fft(signal, title):
        N = len(signal)
        yf = fft(signal)
        xf = fftfreq(N, 1 / fs)[:N // 2]  # Solo parte positiva
        plt.figure(figsize=(10, 4))
        plt.plot(xf, 2.0 / N * np.abs(yf[:N // 2]))
        plt.title(f"FFT de {title}")
        plt.xlabel("Frecuencia (Hz)")
        plt.ylabel("Amplitud")
        plt.grid()
        plt.xlim(0, 200000)
        plt.show()

    plot_fft(signal_A, "Señal A - Diente de Sierra")
    plot_fft(signal_B, "Señal B - sen + cos")
```
